```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
folder_path = r'C:\\Users\\hifia\\Projects\\Dementia Detection and Classification\\Dataset Study\\CSVs'

# Initialize an empty list to store filenames
files = []

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    # Check if the item is a file (not a subfolder)
    if os.path.isfile(os.path.join(folder_path, filename)):
        # Add the filename to the list
        files.append(filename)
```

```python
print(files)
```

```python
def preprocess(fname):
    df = pd.read_csv(f"CSVs/{fname}")
    
    # Print first 5 rows
    print(df.head())

    # Basic description
    print(df.describe())

    # Print no. of null values per attribute
    print(df.isnull().sum())

    # Remove entries where 1/4th or more of the values are missing
    df = df.dropna(thresh=3 * len(df.columns)//4, axis=0)
    
    # Delete Attributes if 50% missing
    df = df.dropna(thresh=len(df)/2, axis=1)

    # Deleting duplicates
    df = df.drop_duplicates()

    # Impute missing values per attribute with mode
    for column in df.columns:
        if df[column].isnull().any():
            df[column].fillna(df[column].mode()[0], inplace=True)

    # export csvs to a new folder, adjusted csvs.
    df.to_csv(f"Adjusted_CSVs/{fname}", index=False)
```

```python
# Preprocessing the files.
for file in files:
    preprocess(file)
```
